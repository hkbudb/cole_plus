```python
import json
import pandas as pd
import matplotlib.pyplot as plt

cole_plus_file = 'readtest-uniform-cole_plus_archive-30000k-fan4-read-test.json'
cole_file = 'readtest-uniform-cole-30000k-mem450000-read-test.json'
def collect_read_time(file_name):
    l = []
    with open(file_name) as f:
        lines = f.readlines()
        for line in lines:
            line = line.strip()
            js_obj = json.loads(line)
            read_time = js_obj['read_time']/1000 #micro second
            l.append(read_time)
    return l

cole_plus_read_times = collect_read_time(cole_plus_file)
cole_read_times = collect_read_time(cole_file)
data = {}
data["cole"] = cole_read_times
data["cole_plus"] = cole_plus_read_times
df = pd.DataFrame(data)
df.boxplot()
print(df.describe())
plt.show()
```

```python
df.to_csv("read_test.csv",index=False, sep='\t')
```
